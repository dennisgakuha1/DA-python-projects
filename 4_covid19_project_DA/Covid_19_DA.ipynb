**Show the number of Confirmed, Deaths and Recovered cases in each Region.**

```python
data.groupby('Region')[['Confirmed','Deaths','Recovered']].sum()
```

**Show the number of Confirmed, Deaths and Recovered cases in each Region. And sort by confirmed cases**

```python
data.groupby('Region')[['Confirmed','Deaths','Recovered']].sum().sort_values(by='Confirmed', ascending=False) .head(3)
```

```python
data.groupby('Region')[['Confirmed','Deaths','Recovered']].sum().sort_values(by='Deaths', ascending=True) .head(3)
```
