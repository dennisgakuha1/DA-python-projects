**Remove nulls**

```python
car.isnull() #check nulls
```

```python
car.isnull().sum()  #Check nulls per column
```

```python
car[car.isnull().any(axis=1)]  #View rows that contain null values
```

```python
car = car.dropna(how='all') #Remove rows where all columns are blank
```

```python
car[car.isnull().any(axis=1)]  #View rows that contain null values
```

```python
car['Cylinders'] = car['Cylinders'].fillna(car['Cylinders'].mean())  #replace null values with mean of the column
```

```python
car[car.isnull().any(axis=1)]  #View rows that contain null values  #No null values
```
